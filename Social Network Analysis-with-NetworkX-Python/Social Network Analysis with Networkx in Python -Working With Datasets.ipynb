### Analysis
+ Degree of Centrality
- Most Influential 
- Most important connection
- Best betweener
- Shortest Path

```python
# General Connection
nx.degree(frnds_graph)
```

```python
# Degree of Connection 
nx.degree(frnds_graph,"Vittal")
```

```python
nx.degree(frnds_graph,"Prasanna")
```

### Most Influential
+ nx.degree_centrality()

```python
nx.degree_centrality(frnds_graph)
```

```python
sorted(nx.degree_centrality(frnds_graph).values())
```

```python
most_influential = nx.degree_centrality(frnds_graph)
```

```python
for w in sorted(most_influential, key=most_influential.get, reverse=True):
    print(w, most_influential[w])
```

### Most Important Connection
+ nx.eigenvector_centrality

```python
nx.eigenvector_centrality(frnds_graph)
```

```python
most_important_link = nx.eigenvector_centrality(frnds_graph)
```

```python
for w in sorted(most_important_link, key=most_important_link.get, reverse=True):
    print(w, most_important_link[w])
```

### Betweener Centrality
+ Bridge/Connect

```python
plt.figure(figsize=(25,15))
nx.draw_networkx(frnds_graph)
plt.show()
```

```python
nx.betweenness_centrality(frnds_graph)
```

```python
best_connector = nx.betweenness_centrality(frnds_graph)
for w in sorted(best_connector, key=best_connector.get, reverse=True):
    print(w, best_connector[w])
```

#### Building A Subgroup 
+  Retrieve all the nodes connected to a given node within a NetworkX graph 
 - Method 1 Breadth-first search
 - Method 2 nx.node_connected_component(G,n)

```python
group1 = nx.bfs_tree(frnds_graph,"Vittal")
```

```python
group2 = nx.bfs_tree(frnds_graph,"Prasanna")
```

```python
group3 = nx.bfs_tree(frnds_graph,"Shraddha")
```

```python
plt.figure(figsize=(25,15))
nx.draw_networkx(group1)
```

```python
plt.figure(figsize=(25,15))
nx.draw_networkx(group2)
```

```python
plt.figure(figsize=(25,15))
nx.draw_networkx(group3)
```
